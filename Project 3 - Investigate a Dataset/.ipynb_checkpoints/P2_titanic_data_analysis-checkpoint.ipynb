### Background & understanding the data

RMS Titanic was a British passenger liner that sank in the North Atlantic Ocean in the early morning of 15 April 1912, after colliding with an iceberg during her maiden voyage from Southampton to New York City. Of the 2,224 passengers and crew aboard, more than 1,500 died, making it one of the deadliest commercial peacetime maritime disasters in modern history. [Source : Wikipedia]

Understanding the data set is one the most important tasks to carry out before starting any kind of analysis. The dataset is provided in a CSV value with each row representing a passenger and each column representing a particular attribute(like Sex, class, Age etc.) of the passenger. 

In this step I will simply try to understand what are the rows present and what each row mean; and what kind of data values it can have. 

Below is a short description of the data fields. 

| Data Columns        | Meaning                           | Data Values             |
| ------------------- | --------------------------------- | ----------------------- |
| PassengerId         | Passenger Sl. No.                 | integer number from 1   |
| Survived            | Passengers survived               | 0 or 1 -- 1 is survived |
| Pclass              | Ticket Class                      | 1, 2, 3                 |
| Name                | Name of the passenger             | Last Name first         |
| Sex                 | Gender                            | male or female          |
| Age                 | Age of the passenger              | floating number(xx.x)   |
| SibSp               | Number of Siblings/Spouses Aboard | 0 or positive integer   |
| Parch               | Number of Parents/Children Aboard | 0 or positive integer   |
| Ticket              | Ticket Number                     | Alphanumeric            |
| Fare                | Passenger Ticket Fare             | Number                  |
| Cabin               | Cabin of the Passenger            | Alphanumeric            |
| Embarked            | Port of Embarkment                | (C = Cherbourg; Q = Queenstown; S = Southampton) |

### Special notes on data -- copied from Kaggle
###### SPECIAL NOTES:

Pclass is a proxy for socio-economic status (SES)
 - 1st ~ Upper; 2nd ~ Middle; 3rd ~ Lower

Age is in Years; Fractional if Age less than One (1)
 - If the Age is Estimated, it is in the form xx.5

With respect to the family relation variables (i.e. sibsp and parch)
some relations were ignored.  The following are the definitions used
for sibsp and parch.

- Sibling:  Brother, Sister, Stepbrother, or Stepsister of Passenger Aboard Titanic
- Spouse:   Husband or Wife of Passenger Aboard Titanic (Mistresses and Fiances Ignored)
- Parent:   Mother or Father of Passenger Aboard Titanic
- Child:    Son, Daughter, Stepson, or Stepdaughter of Passenger Aboard Titanic

Other family relatives excluded from this study include cousins,
nephews/nieces, aunts/uncles, and in-laws.  Some children travelled
only with a nanny, therefore parch=0 for them.  As well, some
travelled with very close friends or neighbors in a village, however,
the definitions do not support such relations.

### Create a function for Hytothesis testing -- Chi test

```python
# Chi Square Testing
def chi_square_test(data):
    chi2, p_value, dof, expected = chi2_contingency(data)
    print "Chi-Squared Score = " + str(chi2)
    print "Pvalue            = " + str(p_value)
    print "\n"
```

### Split the name & Extract the Salutations

```python
#### Function to split name into Salutation, First Name, Middle Name, Last Name
def split_name(name):
    last_name, fm_name = tuple(name.split('(')[0].split(','))
    fm_name = fm_name.split()
    salutation, first_name, middle_name = '', '', ''
    try:
        salutation = fm_name[0]
        first_name = fm_name[1]
        middle_name = fm_name[2]
    except IndexError:
        pass
    return salutation.strip(), first_name.strip(), middle_name.strip(), last_name.strip()
```

```python
titanic_df['Salutation'], titanic_df['First_Name'], titanic_df['Middle_Name'], titanic_df['Last_Name'] = titanic_df['Name'].apply(split_name).str
titanic_df=titanic_df.drop('Name', axis=1)
titanic_df.head(2)
```

### Hypothesis Testing

H0 : There is no significant impact on survivals of sex, i.e., survivals are independent of sex.

H1 : There is significant impact on survivals of sex, i.e., survivals are not independent of sex.
     

```python
# Chi Square Testing
print "------ Results of Chi-Squared test on Sex to Survival. ------\n"
print "Does Sex have a significant effect on Survival?"

chi_square_test(pivot)
```

### Replace the NAN ages with the most like ages within the title category

There are a number of records with missing data for Age. We will not replace them as 0 because that will make 177 persons as aged 0 & in histogram 0 will have a high frequency corrupting the distribution of Ages. We will also not ignore these records as doing so will result in losing around one-fifth of the records.

The best way to replace ages is to find out the probable age group the passenger belongs to. I will use the salutation of a person as an identifier for the group. Example, Salutation Master. is generally used for Kids, Miss. used for unmarried female etc. I will find out the average ages for each group; and the standard deviation of ages for each group. The missing age will be replaced with a value randomly selected from a 95% interval within the Age & Std. Dev. Assumtion is the age distribution is approximately follow the normal distribution(From the Ages distribution plot).

```python
### Find NAN data for Age
age_NANs = titanic_df[titanic_df['Age'].isnull()].head(5)
age_NANs_index = age_NANs.index.values.tolist()

### Replace Age NANs by the Salutation average age
avg_age_by_salutation = titanic_df.groupby(by = 'Salutation').mean()['Age']
avg_age_by_salutation
```

```python
sd_age_by_salutation = titanic_df.groupby(by = 'Salutation').std()['Age']
sd_age_by_salutation = sd_age_by_salutation.fillna(value=0)
sd_age_by_salutation
```

```python
def replace_missing_ages(title):
    return np.random.randint(low = max(0, avg_age_by_salutation[title] - np.random.random() * sd_age_by_salutation[title]), \
                  high = avg_age_by_salutation[title] + np.random.random() * sd_age_by_salutation[title])
```

```python
## Replace NAN age
titanic_df['Age'] = titanic_df['Age'].add(titanic_df[titanic_df['Age'].isnull()]['Salutation'].apply(replace_missing_ages), fill_value=0)
titanic_df.head()
```

```python
# Plot distribution of Ages after correction
facet = sns.FacetGrid(titanic_df,aspect=4, hue='Survival', palette='husl')
facet.map(sns.kdeplot,'Age',shade= True)
facet.set(xlim=(0, titanic_df['Age'].max()))
facet.add_legend()
plt.title("Distribution of Ages by Survivals")
plt.ylabel("Percent")
show(facet)
```

### Looks like the distribution hump/mode shifted a bit towards right. This is due to the NAN ages are replaced with probable values from particular age intervals.
### And the height also now increased slightly by 5%. We can see it's still following the normal distribution approximately.

### Hypothesis Testing -- passengers traveling with Family has better chance of Surviving as compared to people who travelled alone.

H0 : There is no significant relationship of survivals and traveling with family, i.e., we can say they are independent.
    
H1 : There is a relationship of survivals and traveling with family, i.e., we can say they are not independent.


```python
# Family member count to Survivability
pivot = pd.pivot_table(data = titanic_df[['Survival', 'Family_Members_Count']], 
                       index = 'Survival', columns = ['Family_Members_Count'], aggfunc = len).fillna(0)
pivot
```

```python
pivot = pd.pivot_table(data = titanic_df[['Survival', 'Is_with_family']], 
                       index = 'Survival', columns = ['Is_with_family'], aggfunc = len).fillna(0)
pivot
```

```python
# Chi Square Testing
print "------ Results of Chi-Squared test on Sex to Survival. ------\n"
print "Does Sex have a significant effect on Survival?"

chi_square_test(pivot)
```

### In both the cases the Chi-square scores are high and p-values are very less showing there is a very strong relationship between family member count and survivals, hence we will reject the null.
